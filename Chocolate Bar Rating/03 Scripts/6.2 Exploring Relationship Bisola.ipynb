#### 03. Exploring Relationship

```python
# Create a correlation matrix using pandas
#Select only the numerical columns from the DataFrame to avoid errors
numerical_columns = df_choco.select_dtypes(include=['int', 'float'])

numerical_columns.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(numerical_columns.corr())
plt.show()
```

```python
#current dir
cwd = os.getcwd()
cwd
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(numerical_columns.corr(), fignum=f.number) # type of plot
plt.xticks(range(numerical_columns.shape[1]), numerical_columns.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(numerical_columns.shape[1]), numerical_columns.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize = 14) # add font size
plt.title('Correlation Matrix', fontsize = 14) # add title
plt.show()
```

```python
#Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(8,8))
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
correlation = sns.heatmap(numerical_columns.corr(), annot = True, ax = ax, cmap = "plasma") # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.show()
```

```python
# Create a subset excluding the REF column

sub = numerical_columns[['review_year', 'cocoa_percent', 'Rating']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax, cmap="plasma") # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.show()
```

```python
data = df_choco['cocoa_percent']

plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='blue', edgecolor='black')
plt.title("Distribution of Cocoa Percentage", fontsize=16)
plt.xlabel("Cocoa Percentage", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
top_bean_type = df_choco['bean_type'].value_counts().head(10).index
top_bean_type
```

```python
filtered_df1 = df_choco[df_choco['bean_type'].isin(top_bean_type)]
```

```python
avg_rating = filtered_df1.groupby('bean_type')['Rating'].mean().sort_values(ascending=False)
```

```python
# Create a bar plot with log scale
plt.figure(figsize=(10, 6))
aa = sns.barplot(x=avg_rating.index, y=avg_rating.values)
#plt.yscale('log')  # Set the y-axis scale to logarithmic
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.xlabel('bean_type')
plt.ylabel('Rating')
plt.title('Top 10 Bean Type')
#aa.bar_label(aa.containers[0])
plt.tight_layout()

plt.show()  # Show the plot
```

```python
top_countries = df_choco['company_location'].value_counts().head(10)
```

```python
# Create a bar plot with log scale
plt.figure(figsize=(10, 6))
bp =sns.barplot(x=top_countries.index, y=top_countries.values)
#plt.yscale('log')  # Set the y-axis scale to logarithmic
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.xlabel('company_location')
plt.ylabel('Count')
plt.title('Top 10 Countries where the companies are located')
bp.bar_label(bp.containers[0])
plt.tight_layout()
plt.show()
```
